```python
import os
os.getcwd()
os.chdir("C:\\Users\\euphr\\Documents\\Study\\domain\\RT")

import cv2
import numpy as np


# 閾値
result = []
for t in range(1,255):

# 読み込み画像
    grayimg = cv2.imread("img100101.tif" ,0)
    
# 画像の高さh，幅w
    height,width = grayimg.shape

# 全画素数
    sizeall = grayimg.size


    
#二値化してHsを求める
    Ps = []
    for siro in range(1,t+1): 
        
    # 画像の二値化
        ret,th1 = cv2.threshold(grayimg ,siro ,255 ,cv2.THRESH_BINARY)

    # 画像内の白色部分のピクセル数
        sizewhite = cv2.countNonZero(th1)
    
    # Ps(白色ピクセル数の割合)
        Pss = sizewhite / sizeall
        Ps.append(Pss)
    
# Hs
    Hs = sum([v * np.log(v) if v > 0 else 0 for v in Ps])

    
    
#グレースケールでHnを求める
    Pn = []
    for siron in range (t+1,256):
    # 画像の二値化
        ret,th1 = cv2.threshold(grayimg ,siron ,255 ,cv2.THRESH_BINARY)

    # 画像内の白色部分のピクセル数
        sizewhite = cv2.countNonZero(th1)
    
    # Ps(白色ピクセル数の割合)
        Pns = sizewhite / sizeall
        Pn.append(Pns)

#Hn
    Hnss = sum([(1-n) * np.log(1-n) if n < 1 else 0 for n in Pn])
    Hn = Hs + Hnss


#閾値tでの白色ピクセル数を数える    
    # 画像の二値化
    ret,th1 = cv2.threshold(grayimg ,t ,255 ,cv2.THRESH_BINARY)

    # 画像内の白色部分のピクセル数
    sizewhite = cv2.countNonZero(th1)
    
    # Pt(白色ピクセル数の割合)
    Pt = sizewhite / sizeall


# 情報エントロピーS
    S =  Hs / Pt + (Hnss)/(1 - Pt) + np.log(Pt * (1 - Pt)) if Pt < 1 else 0 
    list = [Hs, Hnss, Hn, Pt, S]
    print(list)
    result.append(S)
    
result
result.index(max(result[40:])) 
```
